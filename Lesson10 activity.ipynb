```python
import matplotlib.pyplot as plt

# Data
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temps = [22, 24, 21, 23, 25, 27, 26]

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(days, temps, marker='o', linestyle='-', color='red')
plt.title('Weekly Temperature Trends')
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8, 9]
test_scores = [55, 60, 65, 65, 70, 70, 75, 76, 80]

# Plotting
plt.scatter(hours_studied, test_scores)
plt.title('Scatter Plot of Hours Studied vs. Test Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Test Scores')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data: random daily temperatures for one year
data = np.random.normal(loc=20.0, scale=5.0, size=365)  # average temp 20, std deviation 5
plt.hist(data, bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Daily Temperatures Over One Year')
plt.xlabel('Temperature (degrees Celsius)')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
salespersons = ['Alice', 'Bob', 'Charlie', 'Diana']
sales = [260, 300, 220, 320]

# Creating bar chart
plt.figure(figsize=(8, 5))
plt.bar(salespersons, sales, color='blue', alpha=0.7)
plt.title('Sales Transactions by Salesperson')
plt.xlabel('Salesperson')
plt.ylabel('Number of Transactions')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
brands = ['Apple', 'Samsung', 'Huawei', 'Xiaomi']
market_share = [30, 30, 20, 20]

# Creating pie chart
plt.figure(figsize=(7, 7))
plt.pie(market_share, labels=brands, autopct='%1.1f%%', startangle=140)
plt.title('Smartphone Market Share')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Random test scores data simulating a skewed distribution
test_scores = np.random.normal(70, 10, 100)
test_scores = np.append(test_scores, [40, 40, 90, 100, 100, 100])  # Adding outliers and some very high scores

# Creating box plot
plt.figure(figsize=(6, 8))
plt.boxplot(test_scores, vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title('Distribution of Test Scores')
plt.ylabel('Test Scores')
plt.grid(True)
plt.show()
```
